### 3. Анализ и мониторинг сетевой активности
#### •	Задача: Разработать инструмент для захвата и анализа трафика (например, аналог Wireshark для базового анализа сетевых пакетов).
#### •	Функционал: Отображение типов пакетов, адресов отправителей и получателей, и анализ потенциально вредоносного трафика.

```python
from scapy.all import sniff, IP, TCP, UDP

# Функция для обработки каждого пакета
def packet_handler(packet):
    # Проверяем, является ли пакет IP-пакетом
    if IP in packet:
        src_ip = packet[IP].src
        dst_ip = packet[IP].dst
        protocol = packet[IP].proto
        
        # Определение типа пакета
        if packet.haslayer(TCP):
            pkt_type = "TCP"
        elif packet.haslayer(UDP):
            pkt_type = "UDP"
        else:
            pkt_type = "Другой"

        print(f"Другой: {pkt_type} | Источник: {src_ip} | Адрес: {dst_ip} | Протокол: {protocol}")
        
        # Пример базового анализа на подозрительный трафик
        if packet.haslayer(TCP) and packet[TCP].flags == "S":  # флаг SYN
            print("Обнаружен потенциально подозрительный пакет (установлен флаг SYN)")

# Захват пакетов в реальном времени
sniff(prn=packet_handler, count=10)
```

### Как видно из полученных результатов, написанная Мной программа соответствует функциональным требованиям задачи:
#### Отображает различные типы пакетов: Тут видны как TCP-пакеты (с протоколом 6), так и UDP-пакеты (с протоколом 17).
#### Показывает адреса отправителей и получателей: В каждом пакете указаны IP-адреса источника и назначения. Например, Источник: 192.168.47.18 и Адрес: 173.194.73.94.
#### Анализирует потенциально вредоносный трафик: Пакеты TCP с установленным SYN-флагом отмечены как потенциально подозрительные.
